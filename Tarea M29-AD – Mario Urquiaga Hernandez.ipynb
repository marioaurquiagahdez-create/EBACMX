```python
# WEB SCRAPING CON BEAUTIFULSOUP (sitio con HTTPS)

# Importar librerías necesarias
import requests
from bs4 import BeautifulSoup

# URL de un sitio seguro para practicar scraping
url = 'https://quotes.toscrape.com'
html = requests.get(url).text

# Parsear el HTML con BeautifulSoup
soup = BeautifulSoup(html, 'html.parser')

# Extraer las citas que aparecen en la página
quotes = soup.find_all('div', class_='quote')

# Generación de un listado simple y ordenado (primeras 5 citas)
for i, q in enumerate(quotes[:5], start=1):
    texto = q.find('span', class_='text').get_text()
    autor = q.find('small', class_='author').get_text()
    print(f"{i}. {texto} - {autor}")
```

```python
#USO DE LIBRERÍA REQUESTS CON RESPUESTAS DE API

import requests

# API de datos de gatos (Cat Facts) - listado simple de 5 datos
url_cat = 'https://catfact.ninja/facts?limit=5'
resp_cat = requests.get(url_cat)
data_cat = resp_cat.json()

print("Datos de gatos:")
for i, item in enumerate(data_cat['data'], start=1):
    print(f"{i}. {item['fact']}")

# API de perros (Dog CEO) - listado simple de 10 razas
url_dog = 'https://dog.ceo/api/breeds/list/all'
resp_dog = requests.get(url_dog)
data_dog = resp_dog.json()

breeds = list(data_dog['message'].keys())
print("\nRazas de perros:")
for i, breed in enumerate(breeds[:10], start=1):
    print(f"{i}. {breed.capitalize()}")
```
