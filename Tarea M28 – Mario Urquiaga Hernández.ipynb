```python
# Manejo de archivo '.xlsx'
import pandas as pd

# Leer archivo
df_books = pd.read_excel('Amazon_top100_bestselling_books_2009to2021.xlsx')

# Primeras 10 filas
print("Primeras 10 filas:")
print(df_books.head(10))

# Últimas 10 filas
print("\nÚltimas 10 filas:")
print(df_books.tail(10))

# Seleccionar columnas y crear nuevo archivo
df_books_new = df_books[['title', 'author', 'price', 'year']]
df_books_new.to_csv('books_selected.csv', index=False)
print("Archivo books_selected.csv creado")
```

```python
#Manejo de archivo '.json'
import json
import pandas as pd

# Leer archivo JSON
with open('GOT subtitles.json', 'r') as f:
    data = json.load(f)

# Convertir a lista para trabajar más fácil
subtitles_list = []
for episode, lines in data.items():
    for line_num, text in lines.items():
        subtitles_list.append({'episode': episode, 'line_number': line_num, 'text': text})

df_got = pd.DataFrame(subtitles_list)

# Primeras 10 filas
print("Primeras 10 filas:")
print(df_got.head(10))

# Últimas 10 filas
print("\nÚltimas 10 filas:")
print(df_got.tail(10))

# Seleccionar columnas y crear nuevo archivo
df_got_new = df_got[['episode', 'text']]
df_got_new.to_csv('got_selected.csv', index=False)
print("Archivo got_selected.csv creado")
```
